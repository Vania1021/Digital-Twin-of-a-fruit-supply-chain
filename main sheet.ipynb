```python
import nbformat
from nbconvert import PythonExporter
from nbconvert.preprocessors import ExecutePreprocessor
from sklearn.metrics import mean_squared_error, classification_report, accuracy_score
import asyncio

asyncio.set_event_loop_policy(asyncio.WindowsSelectorEventLoopPolicy())
```

```python
def execute_notebook(notebook_path):
    with open(notebook_path) as f:
        nb = nbformat.read(f, as_version=4)
        ep = ExecutePreprocessor(timeout=600, kernel_name='python3')
        ep.preprocess(nb, {'metadata': {'path': './'}})
        for cell in nb['cells']:
            if 'outputs' in cell:
                for output in cell['outputs']:
                    if 'text' in output:
                        print(output['text'])
                    elif 'data' in output and 'text/plain' in output['data']:
                        print(output['data']['text/plain'])
       
```

```python
def menu():
    print("Select a model to use for prediction:")
    print("1. Decision Tree (Shipment)")
    print("2. Delivery Status")
    print("3. Late Delivery Risk")
    print("4. Shipping Mode")
    print("5. Exit")
```

```python
def main():
    while True:
        menu()
        choice = input("Enter your choice (1-5): ")
        
        if choice == '1':
            print("Executing Decision Tree (Shipment) Notebook...")
            execute_notebook(r'C:\Users\LENOVO\Ml project digital twin\decision tree shipment.ipynb')
            
        elif choice == '2':
            print("Executing Delivery Status Notebook...")
            execute_notebook(r'C:\Users\LENOVO\Ml project digital twin\delivery status.ipynb')
        
        elif choice == '3':
            print("Executing Late Delivery Risk Notebook...")
            execute_notebook(r'C:\Users\LENOVO\Ml project digital twin\late delivery risk.ipynb')
        elif choice == '4':
            print("Executing Shipping Mode Notebook...")
            execute_notebook(r'C:\Users\LENOVO\Ml project digital twin\shipping mode.ipynb')
        
        elif choice == '5':
            print("Exiting the program.")
            break
        
        else:
            print("Invalid choice, please select again.")

if __name__ == "__main__":
    main()
```
